# Concatenate all Datasets

```python
Comp_dataset = pd.concat([merged_df,df2])
```

```python
Comp_dataset # concat all dataset 1,dataset2 & dataset 3
```

```python
Comp_dataset.isnull().sum()
```

```python
Comp_dataset['instant'].mean()
```

```python
Comp_dataset['instant'].fillna(Comp_dataset['instant'].mean(),inplace=True)
```

```python
Comp_dataset.isnull().sum()
```

```python
Q1 = Comp_dataset['Total rentals'].quantile(0.25)
Q3 = Comp_dataset['Total rentals'].quantile(0.75)


Iqr = Q3 - Q1

min_r = Q1 - 1.5 * Iqr
max_r = Q3 + 1.5 * Iqr

outliers=Comp_dataset.index[(Comp_dataset['Total rentals'] < min_r) | (Comp_dataset['Total rentals'] > max_r)]
```

```python
outliers
```

# Removing outliers

```python
Comp_dataset_no_outliers= Comp_dataset.drop(outliers)
```
